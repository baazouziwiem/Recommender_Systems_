```python
"""
Date   : Nov 11 2017
@author: Archana Neelipalayam Masilamani
Task Description:
Retrieveing the top ten favourite movies of any user 

The dataset used is MovieLens data
"""

import pandas as pd

#Read Movielens data  - ratings.csv

ratings_data = pd.read_csv("ml-latest-small/ratings.csv",
                           index_col=False)
movies_data  = pd.read_csv("ml-latest-small/movies.csv",
                           index_col=False, usecols=[0,1])

#Print top 5 rows of Ratings data
ratings_data.head()
```

```python
#Print top 5 rows of Movies data
movies_data.head()
```

```python
# Join Ratings data with title column of Movies data (Like SQL join - on movieId)
data = ratings_data.merge(movies_data,left_on = 'movieId', right_on = 'movieId')
data.head()
```

```python
#Function to retrieve N favourite movies of any user

def favMovie(N,user): 
    favMovieData = data[data['userId']==user].sort_values('rating',ascending=0)
    topFavMovie =  favMovieData[:N]
    return list(topFavMovie['title'])
    
N = 10
userId = 80
resultMovTitle = favMovie(N,userId)

print("Top ",N, " favourite movies by User ID", userId, ": ")

for res in resultMovTitle:
    print(res)

```
